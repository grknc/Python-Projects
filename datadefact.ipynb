```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
```

```python
data = pd.read_csv("Italy-Squad-DefensiveAction.csv",sep=';' ,encoding='latin-1')
```

```python
datateamdef = data.copy()
```

```python
datasquad = datateamdef["Squad"]
```

```python
plt.figure(figsize=(15,5));
bar_ap = sns.pointplot(data =datateamdef, x='Squad', y='Att3rd-Press',linewidth=0.2,palette="flare",linestyles=["-"])
bar_ap.set_xticklabels(bar_ap.get_xticklabels(), rotation=90)
plt.ylabel("Pressure in Attacking 3rd",fontsize = 12,color="green" );
plt.xlabel("",fontsize = 12 , color = 'green');
plt.show()
```

```python
plt.figure(figsize=(15,10));
sns.scatterplot(data=datateamdef, x="Def3rd-Press",y ="Att3rd-Press",hue="Squad",alpha = 1,ci=95);
plt.title("Comparison of Pressure in Defensive/Attacking 3rd Zones Based on Teams in the Italian Serie A league", fontsize = 14,color='black');
for i in range(datateamdef.shape[0]):
    plt.text(x=datateamdef["Def3rd-Press"][i] + 1 , y=datateamdef["Att3rd-Press"][i] + 1,s = datateamdef["Squad"][i],fontdict=dict(size=10))
plt.xlabel("Pressure in Defensive3rd ",fontsize = 12 , color = 'blue');
plt.ylabel("Pressure in Attacking 3rd",fontsize = 12 , color = 'blue');
plt.show();
```

```python
plt.figure(figsize=(15,10));
sns.scatterplot(data=datateamdef, x="Def3rd-Tkl",y ="Att3rd-Tkl",alpha = 1,ci=95);
for i in range(datateamdef.shape[0]):
    plt.text(x=datateamdef["Def3rd-Tkl"][i], y=datateamdef["Att3rd-Tkl"][i]+0.1,s = datateamdef["Squad"][i],fontdict=dict(size=10))
plt.title("Comparison of Tackles in Defensive/Attacking 3rd Zones Based on Teams in the Italian Serie A league", fontsize = 14,color='black');
plt.xlabel("Tackle in Defensive3rd ",fontsize = 12 , color = 'blue');
plt.ylabel("Tackle in Attacking 3rd",fontsize = 12 , color = 'blue');
plt.show();
```

```python
plt.figure(figsize=(15,10));
sns.scatterplot(data=datateamdef, x="Tkl",y ="Def3rd-Tkl",hue="Squad",alpha = 1);
for i in range(datateamdef.shape[0]):
    plt.text(x=datateamdef["Tkl"][i], y=datateamdef["Def3rd-Tkl"][i]+0.1,s = datateamdef["Squad"][i],fontdict=dict(size=10))
plt.title("Comparison of Tackles/Defensive 3rd Zones Based on Teams in the Italian Serie A league", fontsize = 14,color='black');
plt.xlabel("Total Tackles ",fontsize = 12 , color = 'green');
plt.ylabel("Tackle in Defensive 3rd",fontsize = 12 , color = 'green');
plt.show();
```
